```python
import pandas as pd
from collections import Counter
import numpy as np
import random
```

```python
df = pd.read_csv('/content/data/dbir/VCDB/data/csv/vcdb.csv', low_memory=False)

master_count_row = df.shape[0]
master_count_col = df.shape[1] 
print('rows: ', master_count_row)
print ('columns: ', master_count_col)
```

```python
df.head()
```

```python
columns_list = list(df.columns)
filter_hack = []
filter_mal = []
filter_mis = []

for column in columns_list:
  if column.__contains__('action.hacking.variety'):
    filter_hack.append(column)
  if column.__contains__('action.malware.variety'):
    filter_mal.append(column)
  if column.__contains__('action.misuse.variety'):
    filter_mis.append(column)

print(filter_hack)
print(filter_mal)
print(filter_mis)
```

# Mobile App Determination Testing


```python

# See just rows where asset os is known as specified as a mobile os or mobile phone asset 
mobile_df = df.loc[(df['plus.asset_os.Android'])|(df['plus.asset_os.Apple iOS'])|(df['plus.asset_os.BlackBerry OS'])|(df['plus.asset_os.Windows Phone'])|(df["asset.assets.variety.U - Mobile phone"])]
# mobile_df.loc[:,["summary","notes","asset.assets.variety.U - Mobile phone"]]

count_row = mobile_df.shape[0]
count_col = mobile_df.shape[1] 
print('rows: ', count_row)
print('columns: ', count_col)

mobile_df.head()
```

# Export only the Data of interest for further classification

```python
# Trim main dataframe down to only the columns wanted. 

trimmed_index = ['incident_id', 'summary','reference','notes','attribute.confidentiality.data_disclosure.Yes', 'attribute.confidentiality.data_disclosure.Potentially','action.malware.variety.Capture app data', 'action.malware.variety.Capture stored data', 'action.malware.variety.Packet Sniffer','action.physical.variety.Surveillance', 'action.malware.variety.Spyware/Keylogger','action.misuse.Data Mishandling', 'action.misuse.Possession Abuse','action.hacking.variety.Cryptanalysis','action.error.variety.misdelivery','action.malware.variety.Export data', 'action.malware.variety.Phishing', 'action.error.variety.misdelivery','action.social.variety.Bribery', 'action.social.variety.Spam', 'action.malware.variety.Spam','attribute.confidentiality.data.variety.Personal']
print(len(trimmed_index))
print(len(filter_hack))
print(len(filter_mal))
print(len(filter_mis))
trimmed_index = trimmed_index + filter_hack + filter_mal + filter_mis
print(len(trimmed_index))

trimmed_df = df.loc[:,df.columns.isin(trimmed_index)]


count_row = trimmed_df.shape[0]
count_col = trimmed_df.shape[1] 
print('rows: ', count_row)
print('columns: ', count_col)
```

```python
trimmed_df.head()
```

```python
# Get 60 random rows of mobile related

M_sample_df = mobile_df.sample(n=60, replace=False, random_state = 14)  # using state/seed for random sampling with reproducability 
M_sample_df.head()
M_sample_index = list(M_sample_df.index.values)

print(M_sample_index)
print(len(M_sample_index))

# Get 60 more rows (that aren't the same ones) from overall dataframe
sample_index = []
checksum = 0

random.seed(75) # using seed for random sampling with reproducability 
while checksum < 60:
  test_index = random.randint(0,master_count_row)
  while (str(test_index) in trimmed_index):
    test_index=random.randint(0,master_count_row)
  sample_index.append(test_index)
  checksum += 1
  
print(sample_index)
print(len(sample_index))

random_index = M_sample_index + sample_index
```

```python
# Create combined filtered sampled dataframe
filtered_df = trimmed_df.loc[ trimmed_df.index.intersection(random_index)]

f_count_row = filtered_df.shape[0]
f_count_col = filtered_df.shape[1] 
print('rows: ', f_count_row)
print('columns: ', f_count_col)

filtered_df.head()
```

```python
# Export as csv to import into a sheet for hand classification
filtered_df.to_csv('filtered_privacybreach.csv')
```
